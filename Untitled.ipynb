```python
from nba_api.stats.endpoints import leaguestandings
from nba_api.stats.static import teams
import pandas
import json
import math
import numpy as np
import matplotlib.pyplot as plt
pandas.options.mode.chained_assignment = None  # default='warn'

#pull current record from NBA API
record = leaguestandings.LeagueStandings(league_id = "00", season = "2022-23", season_type = "Regular Season")
df = (record.standings.get_data_frame())

nba_teams = teams.get_teams()

team_names = ['Milwaukee Bucks', 'Charlotte Hornets', 'Detroit Pistons', 'Boston Celtics','Oklahoma City Thunder','Minnesota Timberwolves','Washington Wizards','Indiana Pacers','Golden State Warriors','Chicago Bulls','Dallas Mavericks','New Orleans Pelicans','Brooklyn Nets','Atlanta Hawks','Orlando Magic','New York Knicks','Toronto Raptors','Miami Heat','San Antonio Spurs','Los Angeles Lakers','Sacramento Kings','Memphis Grizzlies','Portland Trailblazers','Philadelphia 76ers','Utah Jazz','Phoenix Suns','Denver Nuggets']
cities = [i.rsplit(' ', 1)[0] for i in team_names]

work_table = df[df['TeamCity'].isin(cities)]
```

```python
#here are a set of defined functions that will make the output more readable. Each is documented independently.

#takes in a list of keys to run against an internal dictionary of bets. Need to update to also take in the dictionary, hard coded for now
def results_table():
        
        key_list = ['Tip','Ty','Nick','Donny']
        final_frame = pandas.DataFrame()
        for person in key_list:
            #get the list of bets made by each person, match to the current standings table based on City name, and limit to the
            # info that we care about for this application
            bets_array = pandas.DataFrame(bets.get(person), columns=['TeamCity', 'OverUnder','Line'])
            bets_array=bets_array.set_index("TeamCity")
            temp_table=work_table.merge(bets_array, left_on='TeamCity', right_on='TeamCity')
            temp_table['Owner'] = person
            frames =(final_frame,temp_table)
            final_frame=pandas.concat(frames)
        return final_frame
def calc_stats (table):
    output['Expected Wins'] = (output['WinPCT']*82).astype(int)
    output['Margin'] = (output['Expected Wins']-output['Line'])
    output['ProgressToLine'] = output['Line']-output['WINS']
    conditions = [
        ((output['OverUnder']=="u") & (output['Line']>output['Expected Wins'])),
        ((output['OverUnder']=="o") & (output['Line']<output['Expected Wins'])),
        ((output['OverUnder']=="u") & (output['Line']<output['Expected Wins'])),
        ((output['OverUnder']=="o") & (output['Line']>output['Expected Wins'])),

    ]

    values = [1,1,0,0]
    values2=[-output['Margin'],output['Margin'],-output['Margin'],output['Margin']]
    output['1IfWinning'] = np.select(conditions, values)
    output['Directional Margin'] = np.select(conditions, values2)
    
    return output

```
